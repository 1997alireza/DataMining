### a)

```python
dataset_adr = 'datasets/svmdata2.csv'
```

```python
train, test = extract_dataset(dataset_adr)
```

```python
plt.scatter(train[:, 0], train[:, 1], c=train[:, 2])
plt.show()
```

As is clear from the plot, the features made a circular relation, so we can't use linear classifier; Thus, we use the k(u, v) = u^2 + v^2 to transform the dataset and use a linear classifier.

### b)

```python
def transform(dataset):
    features = dataset[:, 0: 2]
    features = np.square(features)
    features = sum(np.transpose(features))
    return np.transpose(np.array([features, dataset[:, -1]]))
```

```python
transformed_train = transform(train)
plt.scatter(transformed_train[:, 0], transformed_train[:, 1])
plt.ylabel('Label Value')
plt.show()
```

### c)

```python
classifier = build_classifier(train)
print('Error on original dataset =', calculate_error(classifier, train))
```

### d)

```python
classifier = build_classifier(transformed_train)
print('Error on transformed dataset =', calculate_error(classifier, transformed_train))
```
